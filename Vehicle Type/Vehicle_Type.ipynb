# Vehicle Classification using CNN

This notebook classifies vehicle images into different categories using a CNN.

```

import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import os
```

## Load Dataset

```

# Example public dataset loader (folder-based)
img_size = (128, 128)
batch_size = 32

data_dir = "DATASET_PATH_HERE"  # replace if needed

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=img_size,
    batch_size=batch_size
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=img_size,
    batch_size=batch_size
)
```

## CNN Model

```

model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(128,128,3)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(train_ds.class_names), activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

## Training

```

history = model.fit(train_ds, validation_data=val_ds, epochs=5)
```

## Visualization

```

plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

## Conclusion

The experiment confirms that CNNs are effective for vehicle image classification, which is useful for intelligent transportation and surveillance systems.
